#### Buscar documentos

El método **find()** nos ayuda a encontrar los documentos que estamos buscando.  Este método recibe una entrada de tipo llave-valor como criterio de búsqueda. Si no se envía ningún parámetro nos traerá todos los documentos de la colección.  

```
db.coleccion.find(criterio)
```

Ejemplo:


```
db.instructores.find()
```

Ejemplo:

```
db.instructores.find ({"area": "Redes"})
```
Ejemplo: queremos ver la información de los instructores del Huila de Contenido digital.


```
db.instructores.find({"regional": "Huila", "area": "Contenido digital"})
```

Una variante es el método **findOne()** que retorna un solo documento.




##### Operadores de comparación


*   **\$eq:**  Igual que
*   **\$ne:**  No igual que
*   **\$gt:**  Mayor que
*   **\$gte:** Mayor o igual que
*   **\$lt:**  Menor que
*   **\$lte:** Menor o igual que
*   **\$in:**  Se encuentra en

Ejemplo:  Queremos mostrar la información de los instructores de Tolima, Valle y Huila



```
db.instructores.find({"regional": {$in: ["Tolima", "Valle", "Huila"]}})
```





##### Operadores lógicos

*   **\$and:**  y
*   **\$or:**  o
*   **\$not:**  no

Ejemplo:  Queremos consultar a todos los instructores que no son del área de Software.

```
db.instructores.find({ "area": { $not: {$eq: "Software"}}})
```
Ejemplo: Queremos consultar a los instructores que son del Huila y del área de Software:



```
db.instructores.find({$and: [{"regional": "Huila"},{"area": "Desarrollo de Software"}]})
```

#### Elegir qué información mostrar

El método find() tiene un segundo parámetro opcional a través del cual podemos especificar los "campos" que queremos mostrar en cada documento filtrado.  Este parámetro se llama *Projection* y aunque su nombre no es requerido, sabremos que el segundo parámetro hace referencia a él.

El parámetro projection recibe los nombres de las llaves y un valor que puede ser 0 o 1.  El cero (0) indica que el valor no se debe mostrar y el uno (1) que si lo debe hacer.  

**Nota:** en una misma projection no podemos combinar 1s y 0s excepto con el primer campo que tienen todos los documentos llamado **_id**

Ejemplo:  Queremos ver de todos los instructores solo su nombre y regional.

```
db.instructores.find({}, {"_id": 0, "nombre": 1, "regional": 1})
```



#### Actualizar documentos

Los métodos **updateOne()** y **updateMany()** nos permiten actualizar la información de un documento.

Estos métodos reciben 2 parámetros.  El primero es el criterio de búsqueda y el segundo es el nuevo valor a ser asignado al documento que cumple con el criterio de búsqueda.  La asignación se realiza mediante el operador **$set**.  Miremos


```
db.coleccion.updateOne({campo:valor_buscado}, {$set:{campo:nuevo_valor}})
```
Ejemplo:
Vamos a cambiar la regional del instructor Oscar Saenz de Caldas a Risaralda.

```
db.instructores.updateOne({"nombre": "Oscar Saenz"}, {$set: {"regional": "Risaralda"} })
```
Después de comprobarlo, regresa al instructor a su regional Caldas.

El método **updateMany()** lo utilizamos cuando necesitamos actualizar varios registros que cumplan con el criterio de búsqueda.  Por ejemplo, vamos a actualizar el área llamada "Software" por "Desarrollo de Software"



```
db.instructores.updateMany({"area": "Software"},{$set: {"area": "Desarrollo de Software"}})
```





#### Eliminar documentos

Contamos con los métodos **deleteOne()** y **deleteMany()** para eliminar documentos de una colección.  Estos sólo reciben el criterio de búsqueda.

```
db.coleccion.deleteOne(criterio)
db.coleccion.deleteMany(criterio)
```
Ejemplo:
Vamos a ingresar a un instructor adicional para luego proceder a su eliminación

```
db.instructores.insertOne({"nombre": "Floriberto Florez", "area": "Floristería"})
```
Comprobemos su inserción y procedamos a su eliminación:


```
db.instructores.deleteOne({"nombre": "Floriberto Florez"})
```

De manera similar, una vez estamos conectados con la base de datos, podemos conectarnos con alguna de las colecciones existentes en ella o crear una nueva colección así:

A partir de la conexión con una colección, ya podemos aplicar un conjunto de operaciones sobre ella (insertar, consultar, modificar, eliminar, etc)



### Listar el contenido de una colección


```
import pymongo
conexion = pymongo.MongoClient("mongodb://localhost:27017")
bd = conexion["sena"]
instructores = bd["instructores"]
for inst in instructores.find():
    print (inst)
```

Para hacer filtros
```
import pymongo
conexion = pymongo.MongoClient("mongodb://localhost:27017")
bd = conexion["sena"]
instructores = bd["instructores"]
for inst in instructores.find({"area": "Contenido digital"}):
    print (inst)
```
Como podemos ver, los nombres de los métodos, la estructura y contenido de los parámetros mantiene consistencia con la MongoDB Query API

Ejemplo: Queremos ver solo los nombres de los instructores de la regional Tolima

```
import pymongo
conexion = pymongo.MongoClient("mongodb://localhost:27017")
bd = conexion["sena"]
instructores = bd["instructores"]
for inst in instructores.find({"regional": "Tolima"}, {"_id": 0, "nombre": 1}):
    print (inst)
```

Manteniendo la consistencia con la API, pymongo ofrece los métodos **update_one()** y **update_many()**

Ejemplo:  Vamos a actualizar el área de Contenido digital por **Contenidos Digitales** para todos los instructores de esta área

```
import pymongo
conexion = pymongo.MongoClient("mongodb://localhost:27017")
bd = conexion["sena"]
instructores = bd["instructores"]
instructores.update_many({"area": "Contenido digital"}, {"$set": {"area": "Contenidos Digitales"}})
```



### Eliminar documentos

**delete_one()** y **delete_many()** son los métodos usados para eliminar documentos de una colección.  Para el ejemplo insertaremos un nuevo instructor y procederemos a eliminarlo

```
import pymongo
conexion = pymongo.MongoClient("mongodb://localhost:27017")
bd = conexion["sena"]
instructores = bd["instructores"]
instructores.insert_one({"nombre": "Marcos Pinto","area": "Carpinteria"})
```
Después de comprobar su inserción, vamos a eliminarlo

```
import pymongo
conexion = pymongo.MongoClient("mongodb://localhost:27017")
bd = conexion["sena"]
instructores = bd["instructores"]
instructores.delete_one({"nombre": "Marcos Pinto"})
```
