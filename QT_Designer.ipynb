3.  Crear un archivo llamado **calculadora.py** y en él el siguiente código:


```
from PyQt6 import uic
from PyQt6.QtWidgets import QApplication, QMainWindow
import sys

app = QApplication([])
ventana = QMainWindow()
uic.loadUi("interfaz_calculadora.ui", ventana)
ventana.show()
sys.exit(app.exec())
```
Si lo ejecutas, verás la interfaz gráfica pero aún sin procesar las interacciones del usuario (eventos)


Ahora, modificamos el archivo para incluir el código que gestiona los eventos.  El código del archivo calculadora.py quedaría así:



```
from PyQt6 import uic
from PyQt6.QtWidgets import QApplication, QMainWindow, QMessageBox
import sys

def calcular_resultado():
    num1 = ventana.lineEdit.text()
    num2 = ventana.lineEdit_2.text()
    operacion = ventana.comboBox.currentText()
    expresion = num1+operacion+num2
    resultado = eval(expresion)
    mensaje = QMessageBox()
    mensaje.setWindowTitle("Resultado")
    mensaje.setText(f"El resultado de la operación es: {resultado}")
    mensaje.exec()

app = QApplication([])
ventana = QMainWindow()
uic.loadUi("interfaz_calculadora.ui", ventana)

ventana.pushButton.clicked.connect(calcular_resultado)

ventana.show()
sys.exit(app.exec())
```
