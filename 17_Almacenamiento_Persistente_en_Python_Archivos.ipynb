# Almacenamiento Persistente en Python - Archivos

Hasta ahora, el almacenamiento de los datos en nuestros programas se ha realizado en la memoria RAM del computador, lo cuál hace que estos datos se pierdan al finalizar la ejecución de nuestros programas.  

El almacenamiento persistente permite que los datos que reciben y utilizan nuestros programas puedan ser accedidos en cualquier momento.

**¿Qué mecanismos de almacenamiento persistente podemos utilizar en nuestros programas?**



---



Los archivos han sido y seguirán siendo un mecanismo de almacenamiento persistente ampliamente utilizados por nuestras aplicaciones de software.  Aquí algunas razones:

**Simplicidad**: la persistencia de archivos es una técnica simple y fácil de implementar. No se requiere de herramientas o sistemas adicionales para su implementación, ya que simplemente se guardan los datos en archivos en el sistema de archivos del equipo.

**Portabilidad**: los archivos pueden ser transportados fácilmente de un sistema a otro, lo que los hace portátiles. Esto significa que los datos almacenados en archivos pueden ser compartidos y utilizados por diferentes programas en diferentes plataformas.

**Flexibilidad**: los archivos pueden almacenar datos en una variedad de formatos, lo que los hace muy flexibles. Los datos pueden ser almacenados en formatos de texto plano, binarios o en formatos de intercambio de datos, como JSON o XML.

**Eficiencia**: en general, la persistencia de archivos es muy eficiente en términos de recursos de memoria y procesamiento.

**Accesibilidad**: los archivos son muy accesibles, ya que se pueden leer y escribir fácilmente con las herramientas de programación disponibles en el sistema.

Gestionar archivos en Python es una tarea esencial y por esta razón no requerimos ni la instalación ni la importación de paquetes o módulos por fuera de la biblioteca estándar del lenguaje.  

## La función open

Esta función "curiosamente" permite abrir un archivo para su gestión.  La sintaxis básica de la función open es:



```
open (<archivo> , <modo>)
```
donde *archivo* es la ruta absoluta o relativa del archivo que quiero abrir y *modo* es un caracter que indica lo que se quiere hacer con el archivo.





**Modos de apertura:**

*   Read ("r") (Leer)
*   Append ("a") (Agregar)
*   Write ("w") (Escribir)
*   Create ("x") (Crear)

**Tipos de archivos:**
*   ("t") (Modo texto)
*   ("b") (Modo binario)

Por defecto un archivo que se abra sin expresarle su modo, quedará abierto de la forma **"rt"**

## Lectura de archivos

### El método read()

El método read() del objeto file permite la lectura del contenido de un archivo.  

Crea un nuevo archivo llamado *leer_archivo.py* incluye en él las siguientes líneas:


```
archivo = open ("primer_archivo.txt", "r")
contenido = archivo.read()
print (contenido)
```



Ejecuta tu programa y deberás ver el contenido del archivo en la pantalla (tu programa python ha leído el archivo).

**¿Ves algo extraño en el archivo leído?** Probablemente si el archivo leído contiene caracteres especiales como vocales con tílde, el resultado observado en pantalla puede parecer un poco extraño.  Para corregir esta anomalía, existe el atributo **encoding** que se utiliza al momento de abrir el archivo.  

Incluye este atributo en la primera línea de tu programa


```
archivo = open ("primer_archivo.txt", "r", encoding="utf-8")
```



El método read puede recibir como parámetro un número que representa la cantidad de bytes a ser leídos.  Si el parámetro se omite, se leerán todos los bytes.  Comprueba el resultado si modificas la segunda línea de tu programa de esta manera:

```
contenido = archivo.read(5)
```

**Nota:** Una buena práctica de programación es cerrar los archivos después de utilizarlos, para ello, debemos emplear el método **close()**

```
archivo.close()
```



### El método readline()

Éste método nos permite leer una línea de un archivo.  Miremos:



```
archivo = open ("primer_archivo.txt", "r", encoding="utf-8")
print ("Primera línea:")
linea = archivo.readline()
print (linea)
archivo.close()
```

Cada vez que se invoca al método readline() se lee la siguiente línea del archivo.  De esta manera y con la ayuda de un ciclo, podemos leer parte o todo el contenido de un archivo línea a línea.

```
archivo = open ("primer_archivo.txt", "r", encoding="utf-8")
linea = archivo.readline()
while linea:
  print (linea)
  linea = archivo.readline()
archivo.close()
```


### El método readlines()

Éste método lee todas las líneas de un archivo y las retorna en una lista.  Miremos:



```
archivo = open ("primer_archivo.txt", "r", encoding="utf-8")
lineas = archivo.readlines()
print (type (lineas))
print (lineas)
archivo.close()
```

Una vez capturado todo el contenido de un archivo en una lista, podemos acceder a las posiciones de la lista (líneas del archivo) que deseemos o a todas ellas a través de un ciclo:

```
archivo = open ("primer_archivo.txt", "r", encoding="utf-8")
lineas = archivo.readlines()
print ("primera línea:")
print (lineas[0])
print ("todas las líneas:")
for linea in lineas:
  print (linea)
archivo.close()
```

## Modificación de archivos

Existen dos modos que permiten abrir un archivo para modificar su contenido.  Estos son:

Append -> "a" : Significa que se va a abrir un archivo para agregar información al final.

Write -> "w" : Significa que se va a abrir un archivo para modificar la información actual por la nueva información que se envía.



### El método write()

El método que usamos para escribir información en un archivo existente es el método write.  

A continuación realizaremos la práctica de modificar el archivo de texto *primer_archivo.txt* agregando nuevas líneas en él.

Crea un nuevo archivo llamado *adicionar_archivo.py* y agrega a este archivo el siguiente código:

```
archivo = open ("primer_archivo.txt", "a")
archivo.write ("\n Esta es una nueva línea")
archivo.write ("\n Esta es otra línea")
archivo.close()

```



Notarás que las dos líneas fueron agregadas al final del archivo manteniendo el contenido inicial.

Otra manera de modificar un archivo es reemplazando su contenido actual por nuevo contenido.  Para ello, debemos realizar la apertura del archivo con el modo "w".  Miremos:

Crea un nuevo archivo llamado reemplazar_archivo.py y agrega a este archivo el siguiente código:



```
archivo = open ("primer_archivo.txt", "w")
archivo.write ("Este es el nuevo contenido\nque agregamos al archivo\nlo siento si perdiste información importante\nesta es la diferencia de abrir el archivo con el modo w")
archivo.close()
```
Ejecuta el programa y revisa el contenido del archivo *primer_archivo.txt*



### El método writelines()

Si queremos escribir varias líneas de texto en un archivo, podemos usar el método writelines(), el cual recibe como parámetro una lista de cadenas (string) con los datos que queremos escribir en el archivo.  Miremos:



```
archivo = open ("primer_archivo.txt", "a")
lineas = ["\nlinea 1","\nlinea 2","\nlinea 3","\nlinea 4"]
archivo.writelines (lineas)
archivo.close()
```



### El método flush()

Cuando escribimos información en un archivo con el método write o writelines, los datos se almacenan inicialmente en un buffer interno de memoria.  Esta información del buffer pasa realmente al archivo cuando cerramos el objeto file **archivo.close()** Sin embargo, algunas veces podemos requerir llevar los datos del buffer al archivo antes de cerrarlo, para ello utilizamos el método **flush()**

Crea un nuevo archivo llamado *flushing.py* y en él el siguiente código:


```
archivo = open ("primer_archivo.txt", "a")
archivo.write("agrego y guardo en el archivo")
archivo.write("agrego y guardo de nuevo en el archivo")
```

Ejecuta el programa y verifica su comportamiento.  Verás que no se escribe nada en el archivo porque este no se ha cerrado.

Ahora incluye esta línea despues de cada write:



```
archivo.flush()
```

Verifica el resultado en el archivo.



## Acceso a posiciones específicas de un archivo



### El método seek()

Debido a la gestión de archivos a través de buffer de memoria interno, algunas veces es indispensable ubicarnos en la posición del buffer que necesitamos leer.

Para entenderlo mejor miremos el siguiente ejemplo:

Crea un archivo llamado *mutiple_lectura.py* y en él el siguiente código:



```
archivo = open ("primer_archivo.txt")

print ("Primera lectura:\n")
contenido = archivo.read()
print (contenido)

print ("Segunda lectura:\n")
contenido = archivo.read()
print (contenido)

archivo.close()
```
Ejecuta el programa y observa los resultados.

Pudiste comprobar que el contenido del archivo solo se presentó una vez en la pantalla.  **¿Cuál es el motivo?**



El método **seek()** nos permite ubicarnos en una determinada posición (byte) de un archivo (situar el puntero en esa posición) para continuar con las operaciones a partir de esa posición.  

Para el caso del ejercicio anterior de múltiples lecturas, el problema de no realizar la segunda lectura se soluciona simplemente agregando la siguiente línea antes de volver a leer el contenido



```
archivo.seek(0)
```
Comprueba ahora la segunda lectura del archivo.




### El método tell()

Éste método nos devuelve la posición actual del puntero dentro del archivo.  

Agreguemos un par de líneas al programa anterior para entenderlo:



```
archivo = open ("primer_archivo.txt")
print ("posición puntero antes de primera lectura", archivo.tell())
print ("Primera lectura:\n")
contenido = archivo.read()
print (contenido)
print ("\nposición puntero despues de primera lectura", archivo.tell())
print ("Segunda lectura:\n")
archivo.seek(0)
print ("\nposición puntero despues de seek", archivo.tell())
contenido = archivo.read()
print (contenido)
print ("\nposición puntero despues de segunda lectura", archivo.tell())
archivo.close()
```



## El modo +



## Gestores de contexto - with

Los gestores de contexto son una forma de trabajar con recursos que necesitan ser inicializados y/o liberados correctamente, como los archivos, las conexiones de red, las bases de datos, entre otros.

El uso de gestores de contexto genera un código más seguro, ya que el gestor de contexto se encarga de cerrar o liberar los recursos gestionados de manera automática al salir del ámbito del gestor.

En Python, los gestores de contexto se crean con la palabra reservada **with** y el contexto es todo lo que se encuentra dentro de él.

La sintaxis general de un gestor de contexto es la siguiente:


```
with recurso as alias:
    # código que utiliza el recurso
```
Ejemplo:  Leeremos el archivo usando un gestor de contexto.  Crea un archivo llamado *leer_contexto.py* y en él ingresa el siguiente código:



```
with open("primer_archivo.txt") as archivo:
  contenido = archivo.read()
  print (contenido)
```
Ejecuta y verifica el programa.

Ahora crea otro archivo con el nombre *agregar_contexto* y en él el siguiente código:

```
with open("primer_archivo.txt", "a") as arch:
  arch.write ("\nagregada con gestor de contexto")
  
```
Como puedes ver, en ninguno de los dos casos fue necesario utilizar el método close() para cerrar los archivos ya que el gestor de contexto lo hace por nosotros.

Podemos usar tantos gestores de contexto en nuestros programas como necesitemos.


## Excepciones en la gestión de archivos

El control de excepciones es de gran importancia en los programas que gestionan archivos, ya que estos acceden a recursos externos (los archivos) sobre los cuales no podemos controlar características como su existencia, ubicación, nombre, permisos, etc;

Algunas de las excepciones más comunes a la hora de trabajar con archivos son:



*   **FileNotFoundError**.  Ocurre cuando un archivo o directorio es solicitado pero no existe.
*   **PermissionError**. Ocurre cuando se intenta ejecutar una operación sin los permisos de acceso adecuados.
*   **IsADirectoryError**.  Ocurre cuando una operación de archivos se intenta ejecutar en un directorio.



## Archivos binarios
los archivos binarios contienen datos que no están diseñados para ser legibles directamente por humanos. En lugar de caracteres, los archivos binarios contienen información en forma de bytes.

Los archivos binarios se usan para cualquier tipo de archivo que no se pueda representar como texto por ejemplo: imágenes, audio, video, etc.



Los archivos binarios son útiles cuando:

*   Se necesita almacenar datos complejos, como estructuras de datos complejas o matrices multidimensionales, que no se pueden representar fácilmente como texto.

*  Se necesita almacenar grandes cantidades de datos, ya que los archivos binarios generalmente ocupan menos espacio de almacenamiento que los archivos de texto.

*  Se necesita un acceso rápido a los datos, ya que los archivos binarios pueden ser procesados más rápidamente que los archivos de texto.

Ejemplo.  Almacenamiento de imágenes a través de archivos binarios


```
# Escribir una imagen en un archivo binario
with open('image.jpg', 'rb') as f:
    data = f.read()
    
with open('image.bin', 'wb') as f:
    f.write(data)
    
# Leer una imagen de un archivo binario
with open('image.bin', 'rb') as f:
    data = f.read()
    
with open('image_copy.jpg', 'wb') as f:
    f.write(data)

```
