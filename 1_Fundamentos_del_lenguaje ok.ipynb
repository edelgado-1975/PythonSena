**Slicing de Cadenas**

Podemos tomar parte de una cadena a partir de los índices de sus posiciones internas.

En este caso se reciben dos índices separados por dos puntos **:** El primer índice se refiere a la posición inicial y el segundo a la posición final -1.  En caso de faltar alguno de sus índices, se toma el valor por defecto para dicho índice.

Ejemplos:

````python
cadena="lenguaje de programación python"
print (len(cadena))
print (cadena[0:8])
print (cadena[9:11])
print (cadena[12:24])
print (cadena[25:31])
````

**Comprendiendo los valores por defecto y valores negativos en el slicing**

````python
cadena="lenguaje de programación python"
print (cadena[:3])  #len
print (cadena[16:]) #ramación python
print (cadena[-6:]) #python
print (cadena[-4:]) #thon
````

## Métodos de cadenas
A las cadenas les podemos aplicar una serie de métodos útiles predefinidos en python.  Acá un par de ejemplos:

````python
cadena_base = "mi pobre angelito"

cad1 = cadena_base.upper() #convierte a mayúsculas
print (cad1)

numero_de_o = cadena_base.count("o")   #devuelve el número de veces que se encuentra una subcadena
print (numero_de_o)

print (f"la letra 'o' está {numero_de_o} veces en la palabra/frase '{cadena_base}'")

cad2 = cadena_base.replace("pobre", "dulce") #reemplaza una subcadena por otra en la cadena original
print (cad2)
````

# Lectura de datos
La función **input** permite realizar la lectura de datos de entrada por parte del usuario.  La función input devuelve lo ingresado por el usuario en forma de cadena, si se desea capturar un dato de otro tipo se debe hacer la respectiva conversión

````python
marca = input ("Ingrese la marca del auto:")

print ("La marca ingresada por el usuario es", marca)


precio = int(input("Ingrese el precio:"))
print (f"El auto marca {marca} tiene un precio de ${precio} y con el descuento del 10% le queda en {precio*.9}")


#btw
print ("El auto marca {} tiene un precio de ${:,} y con el descuento del 10% le queda en ${:,}".format(marca,precio,precio*0.9))
````

1. Realice un programa que calcule el índice de cosecha de un cultivo en función de la cantidad de frutos recolectados y la cantidad de frutos producidos en total.

Fórmula:
```
Índice de cosecha = (cantidad de frutos recolectados / cantidad de frutos producidos) * 100%
```




````python
cantfrurec= int(input("Digite la cantidad de frutos recolectados: "))
cantfrupro= int(input("Digite la cantidad de frutos producidos: "))
indcos = (cantfrurec / cantfrupro) * 100

print (f"El indice de cosecha: {indcos}")
print ("El indice de cosecha:",format(indcos, '0.3f'),"%")
````

3.	Un alumno desea saber cuál será su calificación final en la materia de Matemáticas. Dicha calificación se compone de los siguientes porcentajes:
55% del promedio de sus tres calificaciones parciales (se debe leer cada calificación parcial).
30% de la calificación del examen final.
15% de la calificación de un trabajo final.


````python
calpar1 = float (input ("Digite la primera calificacion parcial: "))
calpar2 = float (input ("Digite la segunda calificacion parcial: "))
calpar3 = float (input ("Digite la tercera calificacion parcial: "))

parc55 = ((((calpar1 + calpar2 + calpar3)/3)*55)/100)

calexfinal = float (input ("Digite la calificacion del examen final: "))
parc30 = (calexfinal * 30)/100

caltrafinal = float (input ("Digite la calificacion tabajo final: "))
parc15 = (caltrafinal * 15)/100

print ("Nota parcial del 55% es: ",parc55)
print ("Nota parcial del 30% es: ",parc30)
print ("Nota parcial del 15% es: ",parc15)
print ("La Nota final sumando el 55% + 30% + 15% es: ", parc55+parc30+parc15)
````

````python
texto = input("Digite una frase con varias palabras: ")
#metodo 1
numesp = texto.count(" ")
palabras= numesp + 1
print (f"el espacio en blanco ' ' está: {numesp} veces en la frase: '{texto}, por lo cual tiene :{palabras} palabras' ")

#metodo 2
frase = texto.split()
print ("Cantidad de frases", frase)
print ("Cantidad de frases", len(frase))

#metodo 3
frase2 = len(texto.strip().split(" "))
print ("Cantidad de frases", frase2)
````
